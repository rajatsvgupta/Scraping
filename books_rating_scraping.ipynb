Problem Statement:
Loop over book website and get details of all books (book title, Website page no. and rating of the book) and save details in csv

```python
import requests
from bs4 import BeautifulSoup as bs
import lxml
```

```python
i=1
z=0
a=[]
b=[]
c=[]
d=[]
e=[]

z=0
import csv
with open("books_rating_scrap.csv","w",encoding="UTF-8") as f:
    csv_writer=csv.writer(f)
    csv_writer.writerow(["Book Name","Page No.","Rating"])
    while(True):
        x=requests.get(f"http://books.toscrape.com/catalogue/page-{i}.html")
        i=i+1
        y=bs(x.content,"lxml")
        if (len(y.select(".next"))==0):
            z=z+1
        if (len(y.select(".next"))!=0 or z!=2):
            t=y.select(".product_pod")
            for m in range(len(t)):
                if(len(t[m].select(".star-rating.One"))!=0):
                    a.append(t[m].select("a")[1]["title"])
                    csv_writer.writerow([t[m].select("a")[1]["title"],i-1,"One"]) 
                if(len(t[m].select(".star-rating.Two"))!=0):
                    b.append(t[m].select("a")[1]["title"])
                    csv_writer.writerow([t[m].select("a")[1]["title"],i-1,"Two"])
                if(len(t[m].select(".star-rating.Three"))!=0):
                    c.append(t[m].select("a")[1]["title"])
                    csv_writer.writerow([t[m].select("a")[1]["title"],i-1,"Three"])
                if(len(t[m].select(".star-rating.Four"))!=0):
                    d.append(t[m].select("a")[1]["title"])
                    csv_writer.writerow([t[m].select("a")[1]["title"],i-1,"Four"])
                if(len(t[m].select(".star-rating.Five"))!=0):
                    e.append(t[m].select("a")[1]["title"])
                    csv_writer.writerow([t[m].select("a")[1]["title"],i-1,"Five"])
                    
    
        else:
            break
        
    
    
```
